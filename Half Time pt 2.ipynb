```python
def winner(matchid):
    df = sb.events(match_id = matchid)
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    hdf = df[((df['shot_outcome'] == 'Goal') | (df['type'] == 'Own Goal For')) & (df['period'] <= 2)]
    results = {}
    for num, team in tt.items():
        wt = hdf[hdf['team'] == team]
        results[num] = wt.shape[0]
    if results.get('1') == results.get('2'):
        return '0'
    else:
        return max(results, key=results.get)
```

```python
def dueldribblestats(df, period):
    #df = sb.events(match_id = matchid)
    results = {}
    hdf = df[(df['period'] == period)]
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    for num, team in tt.items():
        pdf = hdf[hdf['team'] == team]
        #dw = pdf[pdf['duel_outcome'] == 'Won'].shape[0]
        ad = len(pdf['dribble_outcome'].dropna())
        cd = pdf[pdf['dribble_outcome'] == "Complete"].shape[0]
        tackles = pdf[pdf['duel_type'] == "Tackle"].shape[0]
        ints = len(pdf['interception_outcome'].dropna())
        #results[num + " Duels Won"] = dw
        results[num + " Dribbles Attempted"] = ad
        results[num + " Dribbles Succeeded"] = cd
        results[num + " Tackles Attempted"] = tackles
        results[num + " Interceptions"] = ints
    return results
```

```python
def passingstats(df, period):
    #df = sb.events(match_id = matchid)
    results = {}
    hdf = df[df['period'] == period]
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    tp = hdf[hdf['type'] == "Pass"].shape[0]
    for num, team in tt.items():
        pdf = hdf[(hdf['team'] == team) & (hdf['type'] == "Pass")]
        ttp = pdf.shape[0]
        cp = pdf[pdf['pass_outcome'].isnull()].shape[0]
        off = pdf[pdf['pass_outcome'] == 'Pass Offside'].shape[0]
        corners = pdf[pdf['pass_type'] == "Corner"].shape[0]
        longballs = pdf[pdf['pass_length'] > 37].shape[0]
        slb = pdf[(pdf['pass_length'] > 37) & pdf['pass_outcome'].isnull()].shape[0]
        #crosses = pdf[(pdf['pass_cross'] == True)].shape[0]
        throws = pdf[pdf['pass_type'] == "Throw-in"].shape[0]
        results[num + " Total Passes"] = ttp
        results[num + " Offsides"] = off
        results[num + " Pass Completion"] = round(cp/ttp,3)
        results[num + " Possession"] = round(ttp/tp, 3)
        results[num + " Corners"] = corners
        results[num + " Long Balls"] = longballs
        results[num + " Accurate Long Balls"] = slb
        #results[num + " Crosses"] = crosses
        results[num + " Throws"] = throws
    return results
```

```python
def disciplinestats(df, period):
    #df = sb.events(match_id = matchid)
    results = {}
    hdf = df[(df['period'] == period)]
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    for num, team in tt.items():
        pdf = hdf[hdf['team'] == team]
        try:
            yc = pdf[pdf['foul_committed_card'] == "Yellow Card"].shape[0]
            rc = pdf[(pdf['foul_committed_card'] == "Red Card") | (pdf['foul_committed_card'] == "Second Yellow")].shape[0]
        except:
            yc = 0
            rc = 0
        #fw = pdf.filter(regex='foul_won').dropna(axis = 0, how = 'all').shape[0]
        #fc = pdf.filter(regex= r'^(?=.*foul_committed)(?!.*type)(?!.*card)').dropna(axis = 0, how = 'all').shape[0]
        fouls = pdf[pdf['type'] == 'Foul Committed'].shape[0]
        results[num + " Yellow Cards"] = yc
        results[num + " Red Cards"] = rc
        #results[num + " Fouls Wons"] = fw
        #results[num + " Fouls Committed"] = fc
        results[num + " Fouls Conceded"] = fouls
    return results
```

```python
def shotstats(df, period):
    #df = sb.events(match_id = matchid)
    results = {}
    hdf = df[(df['period'] == period)]
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    for num, team in tt.items():
        pdf = hdf[hdf['team'] == team]
        xg = pdf['shot_statsbomb_xg'].dropna().sum()
        sot = pdf[(pdf['shot_outcome'] == 'Saved') | (pdf['shot_outcome'] == 'Saved To Post')
                 | (pdf['shot_outcome'] == 'Goal')].shape[0]
        goal = pdf[(pdf['shot_outcome'] == 'Goal') | (pdf['type'] == 'Own Goal For')].shape[0]
        shots = pdf['shot_outcome'].dropna().count()
        ww = pdf[pdf['shot_outcome'] == 'Post'].shape[0]
        blocked = pdf[(pdf['shot_outcome'] == 'Blocked')].shape[0]
        results[num + " Total xG"] = round(xg, 3)
        results[num + ' Total Shots on Target'] = sot
        results[num + ' Goals'] = goal
        results[num + ' Shots'] = shots
        results[num + ' Shots Blocked'] = blocked
        results[num + ' Shots woodwork'] = ww
    return results
```

```python
def stats(matchid, period):
    df = sb.events(match_id = matchid)
    d = {**shotstats(df, period), **disciplinestats(df, period), **passingstats(df, period),
        **dueldribblestats(df, period)}
    od = dict(collections.OrderedDict(sorted(d.items())))
    return od
```

```python
data = []
for match in matchids:
    try:
        fhs = stats(match, 1)
        data.append(fhs)
    except:
        pass
```

```python
winners = []
for match in matchids:
    try:
        win = winner(match)
        winners.append(win)
    except:
        pass
```

```python
test = data
fd = collections.defaultdict(list)

for d in test:
    for key, value in d.items():
        fd[key].append(value)
```

```python
df = pd.DataFrame(fd)
df['winner'] = winners
```

```python
def hwin(s1, s2):
    if s1 > s2: 
        return 1
    elif s2 > s1:
        return 2
    else:
        return 0
```

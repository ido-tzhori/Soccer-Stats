```python
def passingstats(df, period):
    results = {}
    hdf = df[df['period'] == period]
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    tp = hdf.shape[0]
    for num, team in tt.items():
        pdf = hdf[(hdf['team'] == team) & (hdf['type'] == "Pass")]
        ttp = pdf.shape[0]
        cp = pdf[pdf['pass_outcome'].isnull()].shape[0]
        off = pdf[pdf['pass_outcome'] == 'Pass Offside'].shape[0]
        corners = pdf[pdf['pass_type'] == "Corner"].shape[0]
        upress = pdf[pdf['under_pressure'] == True].shape[0]
        results[num + " Total Passes"] = [ttp]
        results[num + " Offsides"] = [off]
        results[num + " Pass Completion"] = [round(cp/ttp,3)]
        results[num + " Possesion"] = [round(ttp/tp, 3)]
        results[num + " Corners"] = [corners]
        results[num + " Under Pressure"] = [round(upress/ttp, 3)]
    return results
```

```python
def disciplinestats(df, period):
    results = {}
    hdf = df[(df['period'] == period)]
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    for num, team in tt.items():
        pdf = hdf[hdf['team'] == team]
        try:
            yc = pdf[pdf['foul_committed_card'] == "Yellow Card"].shape[0]
            rc = pdf[(pdf['foul_committed_card'] == "Red Card") | (pdf['foul_committed_card'] == "Second Yellow")].shape[0]
        except:
            yc = 0
            rc = 0
        fw = pdf.filter(regex='foul_won').dropna(axis = 0, how = 'all').shape[0]
        fc = pdf.filter(regex= r'^(?=.*foul_committed)(?!.*type)(?!.*card)').dropna(axis = 0, how = 'all').shape[0]
        results[num + " Yellow Cards"] = [yc]
        results[num + " Red Cards"] = [rc] 
        results[num + " Fouls Wons"] = [fw]
        results[num + " Fouls Committed"] = [fc]
    return results
```

```python
def shotstats(df, period):
    results = {}
    hdf = df[(df['period'] == period)]
    tt = df.team.unique()
    tt = {'1':tt[0],'2':tt[1]}
    for num, team in tt.items():
        pdf = hdf[hdf['team'] == team]
        xg = pdf['shot_statsbomb_xg'].dropna().sum()
        sot = pdf[(pdf['shot_outcome'] == 'Saved') | (pdf['shot_outcome'] == 'Saved To Post')
                 | (pdf['shot_outcome'] == 'Post')].shape[0]
        goal = pdf[(pdf['shot_outcome'] == 'Goal')].shape[0]
        attempts = pdf['shot_outcome'].dropna().count()
        blocked = pdf[(pdf['shot_outcome'] == 'Blocked')].shape[0]
        results[num + " Total xG"] = [xg]
        results[num + ' Total Shots on Target'] = [sot]
        results[num + ' Goals'] = [goal]
        results[num + ' Shot Attempts'] = [attempts]
        results[num + ' Shots Blocked'] = [blocked]
    return results
```

```python
def stats(df, period):
    d = {**shotstats(df, period), **disciplinestats(df, period), **passingstats(df, period)}
    od = collections.OrderedDict(sorted(d.items()))
    return dict(od)
```
